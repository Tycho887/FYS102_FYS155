Konstanter:

- $d=0.96\text{ m}$
- $q = 2.0\times 10^{-6}\text{ C}$

Elektrisk potensial er gitt ved

$V_P = \frac{1}{4\pi\epsilon_0}\sum_{i=1}^{N}{\frac{q_i}{r}}$

Setter inn verdier

$$V_P = \frac{1}{4\pi\epsilon_0}\sum_{i=1}^{N}{\frac{q_i}{r}} = \frac{1}{4\pi\epsilon_0}\left(\frac{-q}{d} + \frac{-q}{d} + \frac{q}{d} + \frac{q}{2d}\right)$$
$$V_P = \frac{1}{4\pi\epsilon_0}
\left(
\frac{-2.0\mu\text{C}}{0.96\text{ m}} + 
\frac{-2.0\mu\text{C}}{0.96\text{ m}} +
\frac{2.0\mu\text{C}}{0.96\text{ m}} +
\frac{2.0\mu\text{C}}{1.92\text{ m}}
\right)$$

$$V_P = -9.4\text{kV}$$
