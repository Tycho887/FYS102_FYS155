```python
import numpy as np
k = 8.988e9
q = 5.0e-6
qA = -q
qB = q
qC = q

distance = 0.03

pos_A = distance * np.array([0, 0])
pos_B = distance * np.array([np.cos(np.pi/3), np.sin(np.pi/3)])
pos_C = distance * np.array([1, 0])

def force(q1, q2, pos1, pos2):
    r = pos2 - pos1
    return k * q1 * q2 * r / np.linalg.norm(r)**3

force_A = force(qA, qB, pos_A, pos_B) + force(qA, qC, pos_A, pos_C)
force_B = force(qB, qA, pos_B, pos_A) + force(qB, qC, pos_B, pos_C)
force_C = force(qC, qA, pos_C, pos_A) + force(qC, qB, pos_C, pos_B)

norm = lambda v: np.linalg.norm(v)

print(f"Magnitude of force on A:  {norm(force_A):.2e}")
print(f"Magnitude of force on B:  {norm(force_B):.2e}")
print(f"Magnitude of force on C:  {norm(force_C):.2e}")
print(f"Net force:  {norm(force_A+force_B+force_C):.2e}")

import matplotlib.pyplot as plt

plt.scatter(*zip(pos_A, pos_B, pos_C))
```
